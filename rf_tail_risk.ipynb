```python
print("="*70)
print("IMPROVED RANDOM FOREST - TAIL-RISK PREDICTION")
print("="*70)

# ============================================
# STEP 1: LOAD DATA & CREATE TARGETS
# ============================================
print("\n[1/7] Loading and preparing data...")

train = pd.read_csv('train_data.csv')
val = pd.read_csv('val_data.csv')
test = pd.read_csv('test_data.csv')

# Check if Tail_Risk_Event exists, if not create it
if 'Tail_Risk_Event' not in train.columns:
    print("  ‚ö†Ô∏è  Tail_Risk_Event not found - creating it now...")

    # Create Next_Day_Return
    train['Next_Day_Return'] = train.groupby('Coin')['Returns'].shift(-1)
    val['Next_Day_Return'] = val.groupby('Coin')['Returns'].shift(-1)
    test['Next_Day_Return'] = test.groupby('Coin')['Returns'].shift(-1)

    # Create Tail_Risk_Event (bottom 5% of returns)
    tail_threshold_train = train.groupby('Coin')['Next_Day_Return'].transform(lambda x: x.quantile(0.05))
    tail_threshold_val = val.groupby('Coin')['Next_Day_Return'].transform(lambda x: x.quantile(0.05))
    tail_threshold_test = test.groupby('Coin')['Next_Day_Return'].transform(lambda x: x.quantile(0.05))

    train['Tail_Risk_Event'] = (train['Next_Day_Return'] < tail_threshold_train).astype(int)
    val['Tail_Risk_Event'] = (val['Next_Day_Return'] < tail_threshold_val).astype(int)
    test['Tail_Risk_Event'] = (test['Next_Day_Return'] < tail_threshold_test).astype(int)

    # Drop rows with NaN in target
    train = train.dropna(subset=['Tail_Risk_Event'])
    val = val.dropna(subset=['Tail_Risk_Event'])
    test = test.dropna(subset=['Tail_Risk_Event'])

    print("  ‚úì Tail_Risk_Event created!")
else:
    print("  ‚úì Tail_Risk_Event already exists!")

# Define all 35 features
feature_cols = [
    'Returns', 'Vol_5d', 'Vol_10d', 'Vol_20d', 'Vol_30d',
    'SMA_5', 'SMA_20', 'SMA_50', 'SMA_200',
    'Price_to_SMA_20', 'Price_to_SMA_50',
    'RSI_14', 'MACD', 'MACD_Signal', 'MACD_Hist',
    'ROC_10', 'ROC_20',
    'BB_Upper', 'BB_Lower', 'BB_Position',
    'Vol_Ratio_5d', 'Vol_Ratio_10d', 'Vol_Ratio_20d', 'Vol_Change',
    'ATR_14', 'ATR_21',
    'Returns_Lag_1', 'Returns_Lag_2', 'Returns_Lag_3',
    'Returns_Lag_5', 'Returns_Lag_7',
    'Volume_Lag_1', 'Volume_Lag_2', 'Volume_Lag_3', 'Volume_Lag_5'
]

# Prepare X and y
X_train = train[feature_cols]
y_train = train['Tail_Risk_Event']
X_val = val[feature_cols]
y_val = val['Tail_Risk_Event']
X_test = test[feature_cols]
y_test = test['Tail_Risk_Event']

print(f"\n‚úì Data prepared:")
print(f"  Train: {len(X_train):,} samples")
print(f"  Val:   {len(X_val):,} samples")
print(f"  Test:  {len(X_test):,} samples")
print(f"  Features: {len(feature_cols)}")

# ============================================
# STEP 2: ADDRESS CLASS IMBALANCE WITH SMOTE
# ============================================
print("\n[2/7] Resampling data to handle class imbalance...")

n_normal_orig = (y_train == 0).sum()
n_tail_risk_orig = (y_train == 1).sum()

print(f"\nOriginal class distribution:")
print(f"  Normal (0):    {n_normal_orig:,} samples ({n_normal_orig/len(y_train)*100:.1f}%)")
print(f"  Tail-Risk (1): {n_tail_risk_orig:,} samples ({n_tail_risk_orig/len(y_train)*100:.1f}%)")
print(f"  Imbalance ratio: {n_normal_orig/n_tail_risk_orig:.1f}:1")

# Apply SMOTE to create synthetic tail-risk examples
print("\nApplying SMOTE (Synthetic Minority Oversampling)...")
smote = SMOTE(sampling_strategy=0.3, random_state=42)  # Target 30% tail-risk
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

n_normal_new = (y_train_resampled == 0).sum()
n_tail_risk_new = (y_train_resampled == 1).sum()

print(f"\n‚úì After SMOTE:")
print(f"  Normal (0):    {n_normal_new:,} samples ({n_normal_new/len(y_train_resampled)*100:.1f}%)")
print(f"  Tail-Risk (1): {n_tail_risk_new:,} samples ({n_tail_risk_new/len(y_train_resampled)*100:.1f}%)")
print(f"  New ratio: {n_normal_new/n_tail_risk_new:.1f}:1")
print(f"  Added {n_tail_risk_new - n_tail_risk_orig:,} synthetic tail-risk samples")

# ============================================
# STEP 3: TRAIN IMPROVED RANDOM FOREST
# ============================================
print("\n[3/7] Training improved Random Forest model...")
print("This may take 10-20 minutes due to larger dataset and deeper trees...")
print("-"*70)

rf_model = RandomForestClassifier(
    n_estimators=200,              # More trees (was 100)
    max_depth=15,                  # Deeper trees (was 10)
    min_samples_split=10,          # More flexible (was 20)
    min_samples_leaf=5,            # More flexible (was 10)
    max_features='sqrt',
    class_weight={0: 1, 1: 20},   # Still weight tail-risk heavily
    random_state=42,
    n_jobs=-1,
    verbose=1
)

# Train on resampled data
rf_model.fit(X_train_resampled, y_train_resampled)

print("\n‚úì Training complete!")

# ============================================
# STEP 4: OPTIMIZE PREDICTION THRESHOLD
# ============================================
print("\n[4/7] Finding optimal prediction threshold...")

# Get probabilities on validation set
val_proba = rf_model.predict_proba(X_val)[:, 1]

# Try different thresholds to maximize recall while maintaining reasonable precision
print("\nTesting different prediction thresholds:")
print("-"*70)
print(f"{'Threshold':<12} {'Precision':<12} {'Recall':<12} {'F1-Score':<12} {'TP':<8} {'FP':<8}")
print("-"*70)

threshold_results = []

for threshold in [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5]:
    val_pred_thresh = (val_proba >= threshold).astype(int)

    cm_temp = confusion_matrix(y_val, val_pred_thresh)

    precision = precision_score(y_val, val_pred_thresh, zero_division=0)
    recall = recall_score(y_val, val_pred_thresh)
    f1 = f1_score(y_val, val_pred_thresh)

    tp = cm_temp[1, 1]  # True positives (caught tail-risks)
    fp = cm_temp[0, 1]  # False positives (false alarms)

    threshold_results.append({
        'threshold': threshold,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'tp': tp,
        'fp': fp
    })

    print(f"{threshold:<12.2f} {precision:<12.4f} {recall:<12.4f} {f1:<12.4f} {tp:<8d} {fp:<8d}")

# Choose threshold that maximizes recall while keeping precision > 0.12
best_threshold = 0.3  # Default
best_recall = 0

for result in threshold_results:
    if result['precision'] >= 0.12 and result['recall'] > best_recall:
        best_recall = result['recall']
        best_threshold = result['threshold']

print("-"*70)
print(f"‚úì Selected threshold: {best_threshold:.2f}")
print(f"  (Prioritizes catching tail-risks over minimizing false alarms)")

# ============================================
# STEP 5: FINAL EVALUATION WITH BEST THRESHOLD
# ============================================
print("\n[5/7] Final evaluation on validation set...")
print("="*70)

# Use best threshold
val_pred = (val_proba >= best_threshold).astype(int)

# Calculate metrics
accuracy = accuracy_score(y_val, val_pred)
roc_auc = roc_auc_score(y_val, val_proba)
precision = precision_score(y_val, val_pred, zero_division=0)
recall = recall_score(y_val, val_pred)
f1 = f1_score(y_val, val_pred)

print(f"\nüìä IMPROVED VALIDATION RESULTS:")
print(f"{'‚îÄ'*70}")
print(f"  Accuracy:   {accuracy:.4f} ({accuracy*100:.2f}%)")
print(f"  ROC-AUC:    {roc_auc:.4f}")
print(f"  Precision:  {precision:.4f} (of predicted tail-risks, {precision*100:.1f}% are correct)")
print(f"  Recall:     {recall:.4f} ‚≠ê (catches {recall*100:.1f}% of tail-risk events)")
print(f"  F1-Score:   {f1:.4f}")
print(f"  Threshold:  {best_threshold:.2f}")
print(f"{'‚îÄ'*70}")

# Detailed classification report
print(f"\nüìã Classification Report:")
print(classification_report(y_val, val_pred,
                          target_names=['Normal (0)', 'Tail-Risk (1)'],
                          digits=4))

# Confusion Matrix
cm = confusion_matrix(y_val, val_pred)
print(f"\nüìä Confusion Matrix:")
print(f"{'‚îÄ'*50}")
print(f"                    Predicted")
print(f"                Normal  Tail-Risk")
print(f"Actual Normal   {cm[0,0]:6d}  {cm[0,1]:6d}")
print(f"       Tail-Risk{cm[1,0]:6d}  {cm[1,1]:6d}")
print(f"{'‚îÄ'*50}")

# Calculate what we caught
n_tail_risks_total = (y_val == 1).sum()
n_tail_risks_caught = cm[1, 1]
n_tail_risks_missed = cm[1, 0]
n_false_alarms = cm[0, 1]

print(f"\nüìà Tail-Risk Detection Performance:")
print(f"  Total tail-risk events:     {n_tail_risks_total:,}")
print(f"  ‚úì Correctly detected:       {n_tail_risks_caught:,} ({recall*100:.1f}%)")
print(f"  ‚úó Missed:                   {n_tail_risks_missed:,} ({(1-recall)*100:.1f}%)")
print(f"  ‚ö† False alarms:             {n_false_alarms:,}")
print(f"\n  Risk Management Impact:")
print(f"  ‚Üí Model warns traders {recall*100:.0f}% of the time before extreme moves")
print(f"  ‚Üí Prevents exposure to {n_tail_risks_caught} out of {n_tail_risks_total} crashes")

# ============================================
# STEP 6: CREATE VISUALIZATIONS
# ============================================
print("\n[6/7] Creating visualizations...")

plt.style.use('default')
sns.set_palette("husl")

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Visualization 1: Feature Importance
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("  Creating feature importance plot...")

importance_df = pd.DataFrame({
    'Feature': feature_cols,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

plt.figure(figsize=(10, 8))
top_15 = importance_df.head(15)
colors = plt.cm.viridis(np.linspace(0.3, 0.9, len(top_15)))
plt.barh(range(len(top_15)), top_15['Importance'], color=colors)
plt.yticks(range(len(top_15)), top_15['Feature'], fontsize=10)
plt.xlabel('Importance Score', fontsize=11)
plt.ylabel('Features', fontsize=11)
plt.title('Top 15 Most Important Features - Improved Random Forest (Tail-Risk)',
          fontsize=13, fontweight='bold', pad=20)
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('rf_improved_feature_importance.png', dpi=300, bbox_inches='tight')
plt.show()
print("  ‚úì Saved: rf_improved_feature_importance.png")

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Visualization 2: ROC Curve
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("  Creating ROC curve...")

fpr, tpr, _ = roc_curve(y_val, val_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, linewidth=2.5, label=f'Improved RF (AUC = {roc_auc:.4f})', color='#2E86AB')
plt.plot([0, 1], [0, 1], 'k--', linewidth=1.5, label='Random (AUC = 0.5000)', alpha=0.7)
plt.fill_between(fpr, tpr, alpha=0.2, color='#2E86AB')
plt.xlabel('False Positive Rate', fontsize=11)
plt.ylabel('True Positive Rate (Recall)', fontsize=11)
plt.title('ROC Curve - Improved Random Forest (Tail-Risk)', fontsize=13, fontweight='bold', pad=20)
plt.legend(loc='lower right', fontsize=10)
plt.grid(alpha=0.3)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.tight_layout()
plt.savefig('rf_improved_roc_curve.png', dpi=300, bbox_inches='tight')
plt.show()
print("  ‚úì Saved: rf_improved_roc_curve.png")

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Visualization 3: Confusion Matrix
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("  Creating confusion matrix...")

plt.figure(figsize=(7, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Normal', 'Tail-Risk'],
            yticklabels=['Normal', 'Tail-Risk'],
            cbar_kws={'label': 'Count'},
            annot_kws={'fontsize': 14, 'fontweight': 'bold'})
plt.ylabel('Actual Label', fontsize=11, fontweight='bold')
plt.xlabel('Predicted Label', fontsize=11, fontweight='bold')
plt.title(f'Confusion Matrix - Improved RF (Threshold={best_threshold:.2f})',
          fontsize=13, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('rf_improved_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()
print("  ‚úì Saved: rf_improved_confusion_matrix.png")

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Visualization 4: Threshold Analysis
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("  Creating threshold analysis plot...")

thresholds_plot = [r['threshold'] for r in threshold_results]
precisions = [r['precision'] for r in threshold_results]
recalls = [r['recall'] for r in threshold_results]
f1s = [r['f1'] for r in threshold_results]

plt.figure(figsize=(10, 6))
plt.plot(thresholds_plot, precisions, 'o-', linewidth=2, label='Precision', color='#E63946')
plt.plot(thresholds_plot, recalls, 's-', linewidth=2, label='Recall', color='#06A77D')
plt.plot(thresholds_plot, f1s, '^-', linewidth=2, label='F1-Score', color='#F77F00')
plt.axvline(best_threshold, color='black', linestyle='--', linewidth=1.5,
            label=f'Selected Threshold ({best_threshold:.2f})')
plt.xlabel('Prediction Threshold', fontsize=11)
plt.ylabel('Score', fontsize=11)
plt.title('Threshold Optimization Analysis', fontsize=13, fontweight='bold', pad=20)
plt.legend(fontsize=10)
plt.grid(alpha=0.3)
plt.xlim([0.05, 0.55])
plt.ylim([0, 1.0])
plt.tight_layout()
plt.savefig('rf_threshold_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
print("  ‚úì Saved: rf_threshold_analysis.png")

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# Visualization 5: Precision-Recall Curve
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
print("  Creating precision-recall curve...")

precision_curve, recall_curve, _ = precision_recall_curve(y_val, val_proba)
avg_precision = average_precision_score(y_val, val_proba)

plt.figure(figsize=(8, 6))
plt.plot(recall_curve, precision_curve, linewidth=2.5,
         label=f'Improved RF (AP = {avg_precision:.4f})', color='#A23B72')
plt.axhline(y=n_tail_risks_total/len(y_val), color='gray', linestyle='--',
            label=f'Baseline ({n_tail_risks_total/len(y_val):.3f})', alpha=0.7)
plt.xlabel('Recall (True Positive Rate)', fontsize=11)
plt.ylabel('Precision', fontsize=11)
plt.title('Precision-Recall Curve - Improved Random Forest',
          fontsize=13, fontweight='bold', pad=20)
plt.legend(loc='upper right', fontsize=10)
plt.grid(alpha=0.3)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.tight_layout()
plt.savefig('rf_improved_precision_recall.png', dpi=300, bbox_inches='tight')
plt.show()
print("  ‚úì Saved: rf_improved_precision_recall.png")

print("\n‚úì All visualizations created!")

# ============================================
# STEP 7: SAVE MODEL AND RESULTS
# ============================================
print("\n[7/7] Saving model and results...")

# Save model with threshold
model_package = {
    'model': rf_model,
    'threshold': best_threshold,
    'feature_cols': feature_cols
}
joblib.dump(model_package, 'rf_improved_tail_risk_model.pkl')
print("  ‚úì Saved model: rf_improved_tail_risk_model.pkl")

# Save results
results_summary = pd.DataFrame([{
    'Model': 'Random Forest (Improved)',
    'Target': 'Tail_Risk_Event',
    'Accuracy': accuracy,
    'ROC_AUC': roc_auc,
    'Precision': precision,
    'Recall': recall,
    'F1_Score': f1,
    'Threshold': best_threshold,
    'N_Features': len(feature_cols),
    'N_Trees': rf_model.n_estimators,
    'Max_Depth': rf_model.max_depth,
    'Used_SMOTE': 'Yes',
    'SMOTE_Ratio': 0.3
}])
results_summary.to_csv('rf_improved_results.csv', index=False)
print("  ‚úì Saved: rf_improved_results.csv")

# Save feature importance
importance_df.to_csv('rf_improved_feature_importance.csv', index=False)
print("  ‚úì Saved: rf_improved_feature_importance.csv")

# Save threshold analysis
threshold_df = pd.DataFrame(threshold_results)
threshold_df.to_csv('rf_threshold_analysis.csv', index=False)
print("  ‚úì Saved: rf_threshold_analysis.csv")

# ============================================
# DOWNLOAD FILES
# ============================================
print("\n" + "="*70)
print("DOWNLOADING FILES")
print("="*70)

files_to_download = [
    'rf_improved_tail_risk_model.pkl',
    'rf_improved_results.csv',
    'rf_improved_feature_importance.csv',
    'rf_threshold_analysis.csv',
    'rf_improved_feature_importance.png',
    'rf_improved_roc_curve.png',
    'rf_improved_confusion_matrix.png',
    'rf_threshold_analysis.png',
    'rf_improved_precision_recall.png'
]

for filename in files_to_download:
    print(f"  Downloading: {filename}")
    files.download(filename)

print("\n‚úì All files downloaded!")

# FINAL SUMMARY
# ============================================
print("\n" + "="*70)
print("TRAINING COMPLETE!")
print("="*70)

print(f"\nFinal Results (Optimized Threshold = {best_threshold:.2f}):")
print(f"  Accuracy:   {accuracy:.4f}")
print(f"  ROC-AUC:    {roc_auc:.4f}")
print(f"  Precision:  {precision:.4f}")
print(f"  Recall:     {recall:.4f} (KEY METRIC)")
print(f"  F1-Score:   {f1:.4f}")

print(f"\nImprovements vs Original:")
print(f"  Original Recall: ~0.22 (22%)")
print(f"  Improved Recall: {recall:.2f} ({recall*100:.0f}%)")
print(f"  Improvement: +{(recall-0.22)*100:.0f} percentage points")

print(f"\nBusiness Impact:")
print(f"  Model now catches {recall*100:.0f}% of tail-risk events ({n_tail_risks_caught} out of {n_tail_risks_total})")
print(f"  Enables traders to proactively reduce exposure before {recall*100:.0f}% of")
print(f"  extreme market crashes, significantly improving risk management capability")

print(f"\nModel Performance Rating:")
if recall >= 0.70 and roc_auc >= 0.72:
    print(f"  EXCELLENT! Production-ready for risk management")
elif recall >= 0.60 and roc_auc >= 0.70:
    print(f"  GOOD! Useful for portfolio protection")
elif recall >= 0.50:
    print(f"  ACCEPTABLE! Shows predictive power")
else:
    print(f"  Needs further improvement")

print(f"\nTechniques Applied:")
print(f"  - SMOTE oversampling (30% tail-risk target)")
print(f"  - Aggressive class weighting (1:20 ratio)")
print(f"  - Deeper trees (max_depth=15)")
print(f"  - More trees (n_estimators=200)")
print(f"  - Optimized prediction threshold ({best_threshold:.2f})")

print("\n" + "="*70)
print("IMPROVED MODEL READY!")
print("="*70)
```
